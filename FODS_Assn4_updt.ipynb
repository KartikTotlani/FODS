```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#importing these libraries to perfrom normalization
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
#importing student performace dataset
df=pd.read_csv('/content/StudentPerformanceFactors.csv')
```

```python
#dividing and segragating dataset into categorical and numerical columns
categorical_columns = ['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level',
                       'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence',
                       'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']
```

```python
sns.boxplot(data=df['Hours_Studied'], color='blue')
plt.title('Original Hours Studied')
```

```python
numerical_columns = ['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 'Tutoring_Sessions',
                     'Physical_Activity', 'Exam_Score']
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_columns)
```

```python
#transforming numercal columns
scaler = StandardScaler()
df_encoded[numerical_columns] = scaler.fit_transform(df_encoded[numerical_columns])
```

```python
plt.figure(figsize=(15, 10))
```

```python
sns.histplot(df['Hours_Studied'], kde=True, color='blue', bins=5)
plt.title('Normalized Hours Studied')
```

```python

sns.histplot(df['Hours_Studied'], kde=True, color='blue', bins=5)
mean = np.mean(df['Hours_Studied'])
median = np.median(df['Hours_Studied'])
mode = stats.mode(df['Hours_Studied'])

# Access the mode value directly using .mode
plt.title(f'Original Hours Studied\nMean: {mean:.2f}, Median: {median}, Mode: {mode.mode}')
plt.axvline(mean, color='r', linestyle='--', label='Mean')
plt.axvline(median, color='g', linestyle='-', label='Median')
# Access the mode value directly using .mode
plt.axvline(mode.mode, color='b', linestyle=':', label='Mode')
plt.legend()
```

```python
sns.histplot(df['Attendance'], kde=True, color='green', bins=5)
plt.title('Normalized Attendance')
```

```python
sns.histplot(df['Previous_Scores'], kde=True, color='red', bins=5)
plt.title('Normalized Previous Scores')
```

```python
sns.histplot(df['Exam_Score'], kde=True, color='purple', bins=5)
plt.title('Normalized Exam Score')
```
