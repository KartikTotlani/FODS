```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean =100
std_dev=4
size = 100000
```

```python
data = np.random.normal(loc=mean, scale=std_dev, size=size)
```

```python
plt.figure(figsize=(10,6))
plt.hist(data, bins=150, color='blue', edgecolor='black', alpha=0.7)
plt.title('Normal Distribution (Mean=100, Standard Deviation=4)', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```
